# ENCODING CATEGORICAL VALUES

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df_cols=('Workclass','Education','Marital_status','Occupation','Relationship','Race','Sex')
for i in df_cols:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

label encoding done in catogerical value columns

# SPLITTING DATASET INTO DEPENDENT AND INDEPENDENT

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

# SPLITING DATASET INTO TRANING AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# TRAINIG THE DATASET ON LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

# PREDICT THE RESULT

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

# MAKING CONFUSION MATRIX AND CHECKING THE ACCURACY

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# 79 % accuracy in logistic regression lets try other one

# TRANING THE DATASET ON RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
classifier_2=RandomForestClassifier(n_estimators=100, criterion='entropy',random_state=0)
classifier_2.fit(x_train,y_train)
```

# PREDICT THE VALUES

```python
y_pred_2=classifier_2.predict(x_test)
print(np.concatenate((y_pred_2.reshape(len(y_pred_2),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred_2)
print (cm)
accuracy_score(y_test,y_pred_2)
```
